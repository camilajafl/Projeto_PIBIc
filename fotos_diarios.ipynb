```python
import os
import pandas as pd
import fitz
```

```python
pasta = os.getcwd() 
pasta_fotos = os.path.join(pasta, "Transcri__es _di_rios e entrevistas_ e v_deos", "fotos")
```

```python
#salvamento e contagem de imagens por região

pastas_destino = ["SPP", "SPA", "SPU", "BHT", "BHV"]

contagem_imagens = {pasta: 0 for pasta in pastas_destino}

def identificar_pasta_destino(nome_arquivo):
    for pasta in pastas_destino:
        if nome_arquivo.startswith(pasta):
            return pasta
    return None

for nome_arquivo in os.listdir(pasta):
    if nome_arquivo.endswith(".pdf"):
        pasta_destino = identificar_pasta_destino(nome_arquivo)
        
        if pasta_destino is None:
            print(f"Pasta de destino não encontrada para o arquivo: {nome_arquivo}")
            continue

        pasta_salvar_imagens = os.path.join(pasta_fotos, pasta_destino)
        os.makedirs(pasta_salvar_imagens, exist_ok=True)

        doc = fitz.open(os.path.join(pasta, nome_arquivo))

        for i in range(len(doc)):
            pagina = doc.load_page(i)
            imagens = pagina.get_images(full=True)
            
            for img_index, img in enumerate(imagens):
                xref = img[0]
                base_imagem = doc.extract_image(xref)
                imagem_dados = base_imagem["image"]
                imagem_extensao = base_imagem["ext"]

                if imagem_extensao.lower() not in ['jpeg', 'jpg']:
                    print(f"Formato de imagem ignorado: {imagem_extensao}")
                    continue
                
                arquivo_imagem = f"{nome_arquivo[:-4]}_p{i}_img{img_index}.{imagem_extensao}"
                caminho_imagem = os.path.join(pasta_salvar_imagens, arquivo_imagem)
                
                with open(caminho_imagem, "wb") as f:
                    f.write(imagem_dados)
                
                contagem_imagens[pasta_destino] += 1
                
                print(f"Imagem salva em {caminho_imagem}")
        
        doc.close()

for pasta, contagem in contagem_imagens.items():
    print(f"Total de imagens salvas na pasta {pasta}: {contagem}")
```

```python
#contagem de fotos após apagar algumas delas

contagem_imagens = {}

for pasta in pastas_destino:
    caminho_pasta = os.path.join(pasta_fotos, pasta)
    if os.path.exists(caminho_pasta):
        contagem_imagens[pasta] = len(os.listdir(caminho_pasta))
    else:
        contagem_imagens[pasta] = 0  

for pasta, contagem in contagem_imagens.items():
    print(f"Total de itens na pasta {pasta}: {contagem}")
```
