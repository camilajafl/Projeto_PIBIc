## Total de homens e mulheres nos grupos focais

```python
GF_genero_local = {}

local_abrev = {
    'SPA' : 'Anchieta',
    'SPP' : 'Paraisópolis',
    'SPU' : 'União',
    'BHT' : 'Taquaril',
    'BHV' : 'Vitória'
}

for sheet in sheets:
    if sheet != 'PSEUDONYMS':
        dados = pd.read_excel(arquivo_excel, sheet_name=sheet, header=None)
        dados = pd.DataFrame(dados)

        local = dados.iloc[3,2]
        GF = dados.iloc[5,2]
        genero = dados.iloc[10,2]

        chave = f'{local} {genero} {GF}'
        if chave in GF_genero_local:
            GF_genero_local[chave] += 1
        else:
            GF_genero_local[chave] = 1

GF_genero_local = pd.DataFrame(list(GF_genero_local.items()), columns=['Local, genero e GF', 'Quantidade'])

GF_genero_local['Local, genero e GF'] = GF_genero_local['Local, genero e GF'].replace({
    'União de Vila Nov Feminino 2': 'União de Vila Nova Feminino 2',
    'União Feminino 1': 'União de Vila Nova Feminino 1'
})

GF_genero_local = GF_genero_local.groupby('Local, genero e GF').sum().reset_index()

GF_genero_local
```
